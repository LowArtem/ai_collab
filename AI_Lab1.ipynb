```python
gender = pd.get_dummies(data['gender'], drop_first=True)
gender.head(5)
```

```python
school_type = pd.get_dummies(data['school_type'], drop_first=True)
school_type.tail(5)
```

```python
school_setting = pd.get_dummies(data['school_setting'], drop_first=True)
teaching_method = pd.get_dummies(data['teaching_method'], drop_first=True)
lunch = pd.get_dummies(data['lunch'], drop_first=True)

#      
data = pd.concat([data, gender, school_type, school_setting, teaching_method, lunch], axis=1)
data
```

```python
#   
data.drop(['school_setting', 'school_type', 'teaching_method', 'gender', 'lunch'], axis=1, inplace=True)
data.head(10)
```

```python
#  boolean  int
data['general_test'] = data['general_test'].astype(int)
data.head(5)
```

```python
#  
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_scale = ['n_student']
scaler.fit(data[cols_to_scale])

data[cols_to_scale] = scaler.transform(data[cols_to_scale])
data.head(5)
```

```python
#     
import math

shuffled = data.sample(frac=1, random_state=1).reset_index()

secLength = math.ceil(len(shuffled) / 3)

train_data = shuffled[:secLength*2]
test_data = shuffled[secLength*2:]

print(len(train_data), len(test_data))
```

```python
train_data.drop(['index'], axis=1, inplace=True)
test_data.drop(['index'], axis=1, inplace=True)
test_data.head(1)
```

```python
#        

x_train = train_data.drop('general_test', axis=1)
y_train = train_data['general_test']

x_train.head(30)
```

```python
# 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)

y_pred_train = model.predict(x_train)
y_pred_train
```

```python
#  
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_train, y_pred_train)
conf_matrix_df = pd.DataFrame(conf_matrix)

conf_matrix_labels = pd.DataFrame(conf_matrix, columns =
['  ', ' '], index = ['  ', ' '])
conf_matrix_labels
```

```python
#  
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(y_train, y_pred_train)
round(model_accuracy, 3)
```
